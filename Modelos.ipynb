## Selección, Configuración y entrenamiento del modelo

Carga de librerías

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb




```

Importamos los datos

```python
df= pd.read_csv('Titanic_train_test.csv')
```

Separamos los datos de test y train

```python
df=df.drop(columns='Unnamed: 0')
```

```python
dftrain=df.iloc[:891]
dftest=df.iloc[891:]
```

```python
dftest=dftest.drop(columns=['Survived'])
```

Separamos en target y features

```python
y=dftrain['Survived']
```

```python
X=dftrain.drop(columns='Survived')
```

Escalamos los features

```python
scaler=StandardScaler()
Xf=scaler.fit_transform(X)
```

```python
Xf
```

Separación del train a train y set

```python
X_train, X_test, y_train, y_test = train_test_split(Xf, y, test_size=0.2, random_state=42)#
```

#### Regresión Logística

```python
    modelo_RL=LogisticRegression(penalty='l2', dual=True, tol=0.0001, C=1.0, fit_intercept=True,
    intercept_scaling=1, class_weight='balanced', solver='liblinear', max_iter=100) #Modelo de regresi'on Logistica
modelo_RL.fit(X_train, y_train)#Entrenamiento del modelo 
y_predRL = modelo_RL.predict(X_test)#Predicciones de los datos de prueba
```

```python

recall = recall_score(y_test, y_predRL)
f1 = f1_score(y_test, y_predRL)
roc_auc = roc_auc_score(y_test, y_predRL)
conf_matrix = confusion_matrix(y_test, y_predRL)

print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1}')
print(f'ROC-AUC: {roc_auc}')
print('Confusion Matrix:')
print(conf_matrix)
```

#### Support Vector Machine

```python
X_train, X_test, y_train, y_test = train_test_split(Xf, y, test_size=0.2, random_state=42)#
```

```python
modelo_SVC = SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', coef0=0.0, 
shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, 
verbose=False, max_iter=-1, decision_function_shape='ovr', break_ties=False, random_state=None)
modelo_SVC.fit(X_train,y_train)
y_predSV = modelo_SVC.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_predSV)
f1 = f1_score(y_test, y_predSV)
roc_auc = roc_auc_score(y_test, y_predSV)
conf_matrix = confusion_matrix(y_test, y_predSV)

print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1}')
print(f'ROC-AUC: {roc_auc}')
print('Confusion Matrix:')
print(conf_matrix)
```

#### Red Neuronal

```python
X_train, X_test, y_train, y_test = train_test_split(Xf, y, test_size=0.2, random_state=42)#
```

```python
net=MLPClassifier(hidden_layer_sizes=(64,64,128,128,128,128,64,64),
 activation='relu', solver='sgd', alpha=0.0001,
 batch_size='auto', learning_rate='adaptive',
 learning_rate_init=0.001, power_t=0.5, max_iter=5000,
 shuffle=True, random_state=None, tol=0.0001,
 verbose=False, warm_start=False, momentum=0.9,
 nesterovs_momentum=True, early_stopping=False,
 validation_fraction=0.1, beta_1=0.9, beta_2=0.999,
 epsilon=1e-08, n_iter_no_change=50, max_fun=15000)
```

```python
#Entrenamiento
net.fit(X_train, y_train)
print("Done ")
```

```python
# 4. evaluate model
acc_train = net.score(X_train, y_train)
print("Accuracy on train = %0.4f " % acc_train)
acc_test = net.score(X_test, y_test)
print("Accuracy on test = %0.4f " % acc_test)
```

```python
y_predNET = net.predict(X_test)
accuracy = accuracy_score(y_test, y_predNET)
f1 = f1_score(y_test, y_predNET)
roc_auc = roc_auc_score(y_test, y_predNET)
conf_matrix = confusion_matrix(y_test, y_predNET)

print(f'Accuracy: {accuracy}')
print(f'F1 Score: {f1}')
print(f'ROC-AUC: {roc_auc}')
print('Confusion Matrix:')
print(conf_matrix)
```
